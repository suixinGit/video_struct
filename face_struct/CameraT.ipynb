```python
import cv2
import numpy 
import dlib

filepath = './Video2018_0611_145624_noglass.wmv'
wnd = 'OpenCV Video'
#获得视频的格式
#使用opencv自带的VideoCapture()函数定义摄像头对象，其参数0表示第一个摄像头，一般就是笔记本的内建摄像头。
videoCapture = cv2.VideoCapture(filepath)
#获取视频信息
frames_num=videoCapture.get(cv2.CAP_PROP_FRAME_COUNT)
#print(frames_num)
#获得码率及尺寸  
fps = videoCapture.get(cv2.CAP_PROP_FPS)  
size = (int(videoCapture.get(cv2.CAP_PROP_FRAME_WIDTH)),   
        int(videoCapture.get(cv2.CAP_PROP_FRAME_HEIGHT)))
#print(size[0], size[1])
  
#cv2.namedWindow(wnd, flags=0)  
#cv2.resizeWindow(wnd,size[0], size[1])  

#a detector to find the faces
detector = dlib.get_frontal_face_detector()
# a shape predictor to find face landmarks so we can precisely localize the face
predictor = dlib.shape_predictor('./shape_predictor_68_face_landmarks.dat')
#the face recognition model
facerec = dlib.face_recognition_model_v1('./dlib_face_recognition_resnet_model_v1.dat')
# 生成dlib的图像窗口
#win = dlib.image_window()

# 读帧  
success, frame = videoCapture.read()
while success : 
    #显示
    #cv2.imshow(wnd, frame) 
    #win.set_image(frame)
    # 取灰度
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    #win.clear_overlay()
    #实现人脸检测/face detection；
    #利用dlib的特征提取器，进行人脸 矩形框 的提取
    # Ask the detector to find the bounding boxes of each face. The 1 in the
    # second argument indicates that we should upsample the image 1 time. This
    # will make everything bigger and allow us to detect more faces.
    dets = detector(frame_gray, 1)
    print("检测到的人脸数目为: {}".format(len(dets)))
    for i, d in enumerate(dets):
        print("第", i+1, "个人脸的矩形框坐标：","left:", d.left(), "right:", d.right(), "top:", d.top(), "bottom:", d.bottom())
        # 绘制矩阵轮廓
        #win.add_overlay(dets[i])
        # 使用predictor来计算面部轮廓
        shape = predictor(frame, dets[i])
        landmarks = numpy.matrix([[p.x, p.y] for p in shape.parts()])
        
        #绘制68个人脸特征点
        for idx, point in enumerate(landmarks):
            # 68点的坐标
            pos = (point[0, 0], point[0, 1])
            
            # 利用cv2.circle给每个特征点画一个圈，共68个
            cv2.circle(frame, pos, 1, color=(0, 255, 0))
            
            # 利用cv2.putText输出1-68
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(frame, str(idx+1), pos, font, 0.3, (0, 0, 255), 1, cv2.LINE_AA)
        #显示
        cv2.imshow(wnd, frame) 
        # Compute the 128D vector that describes the face in img identified by
        # shape.  In general, if two face descriptor vectors have a Euclidean
        # distance between them less than 0.6 then they are from the same
        # person, otherwise they are from different people. Here we just print
        # the vector to the screen.
        face_descriptor = facerec.compute_face_descriptor(frame, shape)
        print('共选取特征维度为%d维：'%len(face_descriptor),face_descriptor)
        # It should also be noted that you can also call this function like this:
        #  face_descriptor = facerec.compute_face_descriptor(img, shape, 100)
        # The version of the call without the 100 gets 99.13% accuracy on LFW
        # while the version with 100 gets 99.38%.  However, the 100 makes the
        # call 100x slower to execute, so choose whatever version you like.  To
        # explain a little, the 3rd argument tells the code how many times to
        # jitter/resample the image.  When you set it to 100 it executes the
        # face descriptor extraction 100 times on slightly modified versions of
        # the face and returns the average result.  You could also pick a more
        # middle value, such as 10, which is only 10x slower but still gets an
        # LFW accuracy of 99.3%.
        # 绘制面部轮廓
        #win.add_overlay(shape)
    #dets = dlib.get_frontal_face_detector(frame)
    #print(dets)
        
    # 保持图像
    #dlib.hit_enter_to_continue()
    cv2.waitKey(10) #延迟
    success, frame = videoCapture.read() #获取下一帧  
videoCapture.release()
cv2.destroyAllWindows() 
```
